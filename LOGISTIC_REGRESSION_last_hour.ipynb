```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("time_frame.csv")
```

```python
data.iloc[:,3:-1]
```

```python
num_data=data.to_numpy()
```

```python
new=num_data[:,3:-1]
```

```python
new.shape
```

```python
meaan=np.arange(new.shape[0])
lis=[]
```

```python
for i in new:
    lis.append(i.mean())
```

```python
lis
```

```python
for i in range(data.iloc[:,-1].shape[0]):
        if(data.iloc[:,-1][i]==lis[i]):
            data.iloc[:,-1][i]=0
        elif(data.iloc[:,-1][i]>lis[i]):
            data.iloc[:,-1][i]=1
        else:
            data.iloc[:,-1][i]=-1
```

```python
data
```

```python
split=int(0.8*data.shape[0])
train=data.iloc[:split,:]
test=data.iloc[split:,:]
```

```python
train_labels=train.iloc[:,-1]
training=train.iloc[:,:63]
test_labels=test.iloc[:,-1]
testing=test.iloc[:,:63]
training=training.drop(columns=['Timestamp'])
testing=testing.drop(columns=['Timestamp'])
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0).fit(training,train_labels)
```

```python
test_output=clf.predict(testing)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(test_labels, test_output)
```
