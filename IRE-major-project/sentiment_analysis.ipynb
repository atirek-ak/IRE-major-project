```python
import glob
import pandas as pd
import numpy as np
import requests
#import html5lib
from bs4 import BeautifulSoup
import datetime
import re


tweets_path = 'Tweets/bitcoin/'
crypto_name = 'bitcoin' # bitcoin, ethereum, bitcoincash, ripple or litecoin
```

```python

cols = ['Tweets','Sentiment']
lst = []
df1 = pd.DataFrame(lst, columns=cols)
lst_dict = []

for f in glob.glob( str(tweets_path) + crypto_name + '*.txt'):
    print(f)
    df_datawise = pd.read_csv(f, sep='\t', encoding='utf-8', parse_dates=True)
    print(df_datawise.size)
    polarity = 0.00
    neutral = positive = negative = 0
    
    
    df2 = pd.DataFrame(columns=cols, index=range(2))

    for idx,row in df_datawise.iterrows():
        polarity += float(row["polarity"])
#         print(row["Tweets"])
        
        
        
        
        # adding reaction of how people are reacting to find average later
        if (float(row["polarity"]) >= -0.1 and float(row["polarity"]) <= 0.1):  
            polar = 1
        elif (float(row["polarity"]) > 0.1):
            polar = 0
        elif (float(row["polarity"]) <-0.1):
            polar = -1
        lst_dict.append({'Tweets':row["Tweets"], 'Sentiment':polar,})
            
df4 = df1.append(lst_dict)
```

## 2. Model for sentiment analysis

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import RegexpTokenizer
from sklearn.model_selection import train_test_split
token=RegexpTokenizer(r'[a-zA-Z0-9]+')
cv=CountVectorizer(stop_words='english',ngram_range=(1,1),tokenizer=token.tokenize)
text_counts=cv.fit_transform(df4['Tweets'].values.astype('U'))

# print(text_counts.count)
print(df4['Sentiment'].count)
X_train, X_test, y_train, y_test = train_test_split(text_counts,df4['Sentiment'], test_size=0.33, random_state=5)

y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python

from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
nb = MultinomialNB()
nb.fit(X_train, y_train)
y_pred = nb.predict(X_test)

print("accuracy :",metrics.accuracy_score(y_test, y_pred))
```
